## 2. Scrubbing the data 

```python
passenger_ds.head(2)
```

```python
from sklearn.utils import shuffle
passenger_ds = shuffle(passenger_ds, random_state=10)
passenger_ds.head()
```

```python
## convert abbrivation to whole word 
gender_dict = {"남":"남자", "여":"여자"}
to_age = lambda x: int(x) * 10
action_dict = {"핸":"핸드폰보기", "생":"생각중", "토":"토크", "책": "책읽기", "노": "노트북", "잠":"잠자는중"}
dress_code_dict = {"정":"정장", "세":"세미정장", "평":"평상복"}
earphone_dict = {"이":"이어폰:O", "노":"이어폰:X"}
station_dict = {"정":"정자역", "판":"판교역"}

def parsePassengerInfo(info):
    """
    parsing passenger information as a labels 
    
    """
    gender = gender_dict.get(info[0], "남자")
    age = to_age(info[1])
    action = action_dict.get(info[2], "생각중")
    dress = dress_code_dict.get(info[3], "평상복")
    is_earphone = earphone_dict.get(info[4], "이어폰사용안함")

    splited_info = info.split(":")
    exit_station = station_dict.get(splited_info[1], "강남역") if len(splited_info)==2 else "강남역"
    
    return [gender, age, action, dress, is_earphone, exit_station]
```

```python
parsePassengerInfo("여2핸평이")
```

```python
passenger_info_list = []
for day, hour, info_list in passenger_ds[['day','hour','info']].values:
    passenger_info_list.extend([[day, hour, seat_loc] + parsePassengerInfo(info) \
                                for seat_loc, info in enumerate(info_list.split())] )
```

```python
passenger_info_list[:2]
```

```python
passenger_ds = pd.DataFrame(passenger_info_list
                , columns=['day', 'hour', 'seat_loc', '성별', '나이', '행동', '복장', '이어폰여부', '내리는역'])
```

```python
print("측정 일수   \t: %4d 일" % len(passenger_ds.day.unique()))
print("측정 승객수  \t: %4d 명" % passenger_ds.shape[0])
print("측정(출근) 시간\t: %s ~ %s" % (passenger_ds.hour.min(),passenger_ds.hour.max()) )
```

* value grouping 

```python
## seat location 
seat_to_group = {
    0:"좌측:사이드",
    1:"좌측:사이드",
    2:"중앙", 
    3:"중앙", 
    4:"중앙",
    5:"우측:사이드", 
    6:"우측:사이드"
}
passenger_ds['좌석위치'] = passenger_ds['seat_loc'].apply(seat_to_group.get)

## station 
passenger_ds['내리는역'] = passenger_ds['내리는역'].apply( lambda x: "강남역" if x == "강남역" else "판교/정자역")

## age 
age_to_group = {
    10:"청년",
    20:"청년",
    30:"중년",
    40:"중년",
    50:"장년",
    60:"장년"
}
passenger_ds['연령대'] = passenger_ds['나이'].apply(age_to_group.get)
passenger_ds['count'] = 1
```

```python
passenger_ds.to_csv("./resource/passenser.csv", index=None)
```
