```python
## lambda 함수 사용
## 코드의 간결성을 위해 익명함수 형태로 사용

def add_one(x):
    return x+1

add_one_lambda = lambda x :x+1

print(add_one(1), add_one_lambda(1))
```

```python
## apply 함수 설명

ds['나이'].apply(add_one)
```
